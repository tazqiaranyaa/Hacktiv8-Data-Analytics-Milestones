```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# Setup WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')  
chrome_options.add_argument('--start-maximized')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
service = Service("/Users/tazqiaranya/Desktop/bootcamp_data/chromedriver")
driver = webdriver.Chrome(service=service, options=chrome_options)

# Daftar halaman yang ingin di-scrape 
urls = [
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=1",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=2",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=3",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=4"
]

all_data = []

for url in urls:
    driver.get(url)
    time.sleep(2)

    # Scroll bertahap 
    for _ in range(20):  # jumlah scroll bisa disesuaikan
        driver.execute_script("window.scrollBy(0, 1500);")
        time.sleep(1)

    # Ambil konten HTML
    soup = BeautifulSoup(driver.page_source, "html.parser")
    product_cards = soup.find_all("div", class_="product product--v3")

    # Ekstrak data
    for card in product_cards:
        brand = card.find("a", class_="product__brand")
        name = card.find("a", class_="product__name")
        price = card.find("div", class_="product__price product__price_c-danger")
        rating = card.find("span", class_="product__stars")
        review = card.find("span", class_="product__reviews")
        discount = card.find("span", class_="product__info-sale")

        all_data.append({
            "brand": brand.get_text(strip=True) if brand else None,
            "product_name": name.get_text(strip=True) if name else None,
            "price": price.get_text(strip=True) if price else None,
            "rating": rating.get_text(strip=True) if rating else None,
            "review": review.get_text(strip=True).strip("()") if review else "0",
            "discount": discount.get_text(strip=True) if discount else "0%"
        })

# Buat DataFrame 
df = pd.DataFrame(all_data)

# Tampilkan info
print(f"Jumlah total produk: {len(df)}")
print(df.head())

# Tutup browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# Dikarenakan price tidak terekstrak, ekstrak ulang price after dan before discountnya saja
# Setup WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')  
chrome_options.add_argument('--start-maximized')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
service = Service("/Users/tazqiaranya/Desktop/bootcamp_data/chromedriver")
driver = webdriver.Chrome(service=service, options=chrome_options)

# List halaman yang mau diekstrak
urls = [
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=1",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=2",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=3",
    "https://www.sociolla.com/139-makeup?sort=-created_at&products=all_product&rating=5&page=4"
]

# Ekstrak harga
price_data = []

for url in urls:
    driver.get(url)
    time.sleep(3)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    product_cards = soup.find_all("div", class_="product product--v3")

    for card in product_cards:
        price_container = card.find("h2", class_="product__price")
        if price_container:
            spans = price_container.find_all("span")
            
            # 
            if spans:
                price_before = spans[0].get_text(strip=True)
                price_after = price_container.get_text(strip=True).replace(price_before, '').strip()
            else:
                price_after = price_container.get_text(strip=True)
                price_before = None
        else:
            price_after = None
            price_before = None

        price_data.append({
            "price_after_discount": price_after,
            "price_before_discount": price_before
        })

# Simpan ke DataFrame dengan nama df_price
df_price = pd.DataFrame(price_data)
print(df_price.head())

# Tutup browser
driver.quit()
```

```python
# Bersihkan data rating dan langsung simpan ke df_final
#rubah yang none menjadi 5.0, karena dari awal scrape website, sudah dipilih produk dgn rating 5
df_final = df.copy()
df_final["rating"] = (
    df_final["rating"]
    .astype(str)
    .fillna("5.0")
    .str.strip()
    .replace("None", "5.0")
    .astype(float)
)
```

```python
df_final = pd.concat([df, df_price], axis=1)
```

```python
# data cleaning
# Jika discount adalah '0%' dan price_before_discount kosong, isi dengan price_after_discount
df_final.loc[
    (df_final["discount"] == "0%") & (df_final["price_before_discount"].isna()),
    "price_before_discount"
] = df_final["price_after_discount"]
```

```python
# Hapus baris dengan index 5, 16, 22, 25, 28 karena price after discountnya harganya bentuk range (cth. 50.000-60.000), drpd sulit ngerubahnya, sekalian dihapus aja
df_final = df_final.drop(index=[6, 17, 23, 26, 29]).reset_index(drop=True)
```

```python
# data cleaning
# Bersihkan kolom harga dan ubah ke integer
for col in ['price_after_discount', 'price_before_discount']:
    df_final[col] = (
        df_final[col]
        .astype(str)                          # memastikan dalam bentuk string
        .str.replace("Rp", "", regex=False)   # hilangkan 'Rp'
        .str.replace(".", "", regex=False)    # hilangkan titik pemisah ribuan
        .str.strip()                          # hilangkan spasi ekstra
        .replace("None", None)                # ubah 'None' string jadi None
        .replace("", None)                    # hilangkan string kosong
    )
    df_final[col] = df_final[col].astype(float).astype('Int64')  # ubah jadi integer (support NaN)
```

```python
#cleanse data kalau diskonnya "SALE" berubah dengan rumus diskon
# Pastikan kolom harga berupa float
df_final["price_before_discount"] = df_final["price_before_discount"].astype(float)
df_final["price_after_discount"] = df_final["price_after_discount"].astype(float)

# Buat kondisi: jika discount == "SALE", hitung ulang menggunakan rumus
mask_sale = df_final["discount"].str.upper() == "SALE"

# Hitung diskon dalam persen (untuk baris yang SALE)
discount_calculated = (
    (df_final.loc[mask_sale, "price_before_discount"] - df_final.loc[mask_sale, "price_after_discount"]) /
    df_final.loc[mask_sale, "price_before_discount"]
) * 100

# Ubah ke format persen dan isi ke kolom discount
df_final.loc[mask_sale, "discount"] = discount_calculated.round(0).astype(int).astype(str) + "%"
```

```python
#Ganti data float jadi int
df_final["price_after_discount"] = df_final["price_after_discount"].astype("int64")
df_final["price_before_discount"] = df_final["price_before_discount"].astype("int64")
```

```python
#Ganti diskon (str) jadi float
# Hapus tanda % dan ubah ke float
df_final['discount'] = df_final['discount'].str.replace('%', '').astype(float) / 100
```

```python
#tambah kolom baru
makeup_types = [
    "Palette", "Eyeliner", "Mascara", "Brow", "Blush", "Tint", "Lipstick",
    "Lip", "Concealer", "Foundation", "Cushion",
    "Cream", "Contour", "Bronzer", "Highlighter",
    "Powder", "Setting Spray",
    "Primer", "Bundle", "Lash", "Eyebrow", "Set", "Base", "Pack"
]

makeup_types_lower = [m.lower() for m in makeup_types]

def extract_makeup_type(product_name):
    name_lower = product_name.lower()
    for original, lowered in zip(makeup_types, makeup_types_lower):
        if lowered in name_lower:
            return original
    return "Unknown"

df_final['makeup_type'] = df_final['product_name'].apply(extract_makeup_type)
```

```python
#rekategorisasi makeup type
makeup_type_map = {
    "Palette": ["palette"],
    "Eyeliner": ["eyeliner"],
    "Mascara": ["mascara"],
    "Brow": ["brow", "eyebrow"],
    "Blush": ["blush"],
    "Tint": ["tint"],
    "Lip": ["lipstick", "lip"],
    "Concealer": ["concealer"],
    "Complexion": ["foundation", "base", "cushion", "cream"],
    "Contour": ["contour"],
    "Bronzer": ["bronzer"],
    "Highlighter": ["highlighter"],
    "Powder": ["powder"],
    "Setting Spray": ["setting spray"],
    "Primer": ["primer"],
    "Bundle": ["pack", "set", "bundle"],
    "Lash": ["lash"]
}

def extract_makeup_type(product_name):
    name_lower = product_name.lower()
    for category, keywords in makeup_type_map.items():
        if any(keyword in name_lower for keyword in keywords):
            return category
    return "Unknown"

df_final['makeup_type'] = df_final['product_name'].apply(extract_makeup_type)
```
